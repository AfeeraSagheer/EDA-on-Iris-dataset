# **7_Visualization** **BY** **Matplotlib** **and** **Seaborne**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd # Ensure pandas is imported for DataFrame operations

# Reload the full iris dataset to get the target (species) information
iris_data_bunch = load_iris()

# Add the 'species' column to the existing 'iris' DataFrame.
# The 'iris' DataFrame in the kernel state only contains features, so we add the target here.
iris['species'] = iris_data_bunch.target

# Species 0 (setosa)
setosa = iris[iris["species"] == 0]
plt.scatter(
    setosa["sepal length (cm)"],
    setosa["sepal width (cm)"],
    label="setosa"
)

# Species 1 (versicolor)
versicolor = iris[iris["species"] == 1]
plt.scatter(
    versicolor["sepal length (cm)"],
    versicolor["sepal width (cm)"],
    label="versicolor"
)

# Species 2 (virginica)
virginica = iris[iris["species"] == 2]
plt.scatter(
    virginica["sepal length (cm)"],
    virginica["sepal width (cm)"],
    label="virginica"
)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Dataset")
plt.legend()
plt.show()
```

```python
plt.plot(iris["sepal length (cm)"], color = "pink")
plt.title("Sepal Length Line Plot")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.show()
```

```python
plt.hist(iris["petal length (cm)"], bins=20, color= "red")
plt.title("Petal Length Distribution")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(iris["sepal width (cm)"])
plt.title("Sepal Width Boxplot")
plt.ylabel("Values")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the average sepal length for each species
# Assuming 'iris' DataFrame and 'species' column are available from previous cells
avg_sepal_length = iris.groupby('species')['sepal length (cm)'].mean()

# Get species names from the loaded iris data to label the bars correctly
from sklearn.datasets import load_iris
iris_data_bunch = load_iris()
species_names = iris_data_bunch.target_names

# Use the calculated average sepal length for plotting
# The index of avg_sepal_length corresponds to species IDs (0, 1, 2)
# We map these to actual species names using species_names
plt.bar(species_names[avg_sepal_length.index], avg_sepal_length.values, color= "brown")
plt.title("Average Sepal Length per Species")
plt.xlabel("Species")
plt.ylabel("Average Length")
plt.show()
```

```python
# here we are giving two subplots to plot two figures
fig, ax = plt.subplots(1, 2)

# First plot
ax[0].hist(iris["sepal length (cm)"])
ax[0].set_title("Sepal Length")

# Second plot
ax[1].hist(iris["petal length (cm)"])
ax[1].set_title("Petal Length")

plt.show()
```

```python
from sklearn.datasets import load_iris

# Reload the iris dataset object to access target_names
iris_data_bunch = load_iris()
species_names = iris_data_bunch.target_names

counts = iris["species"].value_counts()
# creates pie chart giving percentage of value counts of each speceis
plt.pie(counts, labels=species_names, autopct='%1.1f%%')
plt.title("Species Distribution")
plt.show()
```

```python
# it gives the petal length vs sepal length of all three speceis in iris
import seaborn as sns
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", hue="species", data=iris)
plt.title("Sepal vs Petal Length")
plt.show()
```

```python
# it distributes sepal length in iris
sns.histplot(data=iris, x="sepal length (cm)", kde=True)
plt.title("Distribution of Sepal Length")
sns.set_palette("deep")
plt.show()
```

```python
# petal length by each speceis
sns.boxplot(x="species", y="petal length (cm)", data=iris)
plt.title("Petal Length by Species")
sns.set_palette("bright")
plt.show()
```

```python
# gives the sepal width by each speceis
sns.violinplot(x="species", y="sepal width (cm)", data=iris, color = "green")
plt.title("Sepal Width Distribution")
plt.show()
```

```python
# gives all the lengt and width of speceis
sns.pairplot(iris, hue="species")
sns.set_palette("bright")
plt.show()
```

```python
# it gives the correlation of all speceis
corr = iris.corr(numeric_only=True)
sns.set_palette("coolwarm")

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```
